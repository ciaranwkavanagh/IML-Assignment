```python
import os
import json
```

```python
#Getting the list of files present in the current working directory
notebooks = [file for file in os.listdir(os.getcwd()) if file.endswith('.ipynb')]

notebooks
```

```python
def combine_notebook(notebook1, notebook2, notebook3, combined_file_name):
    # Load the first notebook
    with open(notebook1, mode='r', encoding='utf-8') as f:
        combined = json.load(f)

    # Load the second and extend
    with open(notebook2, mode='r', encoding='utf-8') as f:
        nb2 = json.load(f)
        combined['cells'].extend(nb2['cells'])
        print(f'merging: {notebook1}, {notebook2}')
    
    # Load the third and extend
    with open(notebook3, mode='r', encoding='utf-8') as f:
        nb3 = json.load(f)
        combined['cells'].extend(nb3['cells'])
        print(f'merging: {notebook3}')

    # Write out the combined notebook
    with open(combined_file_name, mode='w', encoding='utf-8') as n:
        json.dump(combined, n)

    print('Generated file: {}'.format(combined_file_name))
    return os.path.realpath(combined_file_name)
```

```python
combine_notebook('Final Code.ipynb','Wessel final.ipynb','IML_Regression_Max_Final.ipynb','IML-Final-Code.ipynb')
```
